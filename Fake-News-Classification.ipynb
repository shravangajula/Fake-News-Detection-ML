```python
# Import necessary libraries
import nltk
import pandas as pd

# Download 'punkt' tokenizer from NLTK
nltk.download('punkt')
```

```python

# Load fake and genuine news datasets
fake = pd.read_csv('Fake.csv')
genuine = pd.read_csv('True.csv')
```

```python
fake
```

```python

# Add a column to each dataset to indicate genuineness: 0 for fake news and 1 for genuine news
fake['genuiness'] = 0
genuine['genuiness'] = 1
```

```python
genuine
```

```python
# Concatenate the two datasets to create a single dataset
dataset = pd.concat([fake,genuine])
```

```python
dataset
```

Pre Processing

```python
# Reset the index of the dataset
dataset = dataset.reset_index(drop=True)
```

```python
dataset
```

```python


# Remove unnecessary columns from the dataset
dataset = dataset.drop(['date', 'title', 'subject'], axis=1)
```

```python
dataset
```

```python
# Check the data type of the 'text' column and convert it to string if necessary
print(dataset['text'].dtype)
dataset['text'] = dataset['text'].astype(str)
```

```python

# Tokenize the text using NLTK's word_tokenize function
from nltk.tokenize import word_tokenize
dataset['text'] = dataset['text'].apply(word_tokenize)
```

```python
dataset
```

```python

# Stem the words using NLTK's SnowballStemmer
from nltk.stem.snowball import SnowballStemmer
sb = SnowballStemmer('english')
def stem_it(text):
  return [sb.stem(word) for word in text]
dataset['text'] = dataset['text'].apply(stem_it)
```

```python
dataset
```

```python
# Remove stop words using a custom function
def stop_removel(text):
  return [word for word in text if len(word) > 2]
dataset['text'] = dataset['text'].apply(stop_removel)
```

```python
dataset
```

```python
# Convert the tokenized text back to a single string
dataset['text'] = dataset['text'].apply(''.join)
```

```python
dataset
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(dataset['text'], dataset['genuiness'], test_size=0.25)
```

```python
x_train
```

```python
y_train
```

```python
# Convert the text into a TF-IDF matrix
from sklearn.feature_extraction.text import TfidfVectorizer
tfidfv = TfidfVectorizer(max_df=0.7)
tfidfv_train = tfidfv.fit_transform(x_train)
tfidfv_test = tfidfv.transform(x_test)
```

```python
print(tfidfv_train)
```

```python

# Train a logistic regression model and make predictions
from sklearn.linear_model import LogisticRegression
ir = LogisticRegression(max_iter=1000)
ir.fit(tfidfv_train, y_train)
y_pred = ir.predict(tfidfv_test)
```

```python
# Calculate accuracy of the logistic regression model
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python

# Train a Passive Aggressive classifier model and make predictions
from sklearn.linear_model import PassiveAggressiveClassifier
pac = PassiveAggressiveClassifier(max_iter=1000)
pac.fit(tfidfv_train, y_train)
y_pred_pac = pac.predict(tfidfv_test)
```

```python
# Calculate accuracy of the Passive Aggressive classifier model
accuracy_score(y_test,y_pred_pac)
```
